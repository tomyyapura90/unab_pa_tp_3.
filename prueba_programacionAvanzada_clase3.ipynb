```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def eje_x(self):
        return self.x

    def eje_y(self):
        return self.y

    def impresion(self):
        return f"({self.x}, {self.y})"

    def opuesto(self):
        return Punto(-self.x, -self.y)
```

```python
class Linea:
    def __init__(self, punto_a, punto_b):
        self._punto_a = punto_a
        self._punto_b = punto_b

    def mueve_derecha(self, distancia):
        self._punto_a.x += distancia
        self._punto_b.x += distancia

    def mueve_izquierda(self, distancia):
        self._punto_a.x -= distancia
        self._punto_b.x -= distancia

    def mueve_arriba(self, distancia):
        self._punto_a.y += distancia
        self._punto_b.y += distancia

    def mueve_abajo(self, distancia):
        self._punto_a.y -= distancia
        self._punto_b.y -= distancia
```

```python
class Cancion:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    def get_titulo(self):
        return self.titulo

    def get_autor(self):
        return self.autor

    def set_titulo(self, nuevo_titulo):
        self.titulo = nuevo_titulo

    def set_autor(self, nuevo_autor):
        self.autor = nuevo_autor
```

```python
class Libro:
    def __init__(self, titulo, autor, isbn, paginas, edicion, editorial, ciudad, pais, fecha_edicion):
        self.titulo = titulo
        self.autor = autor
        self.isbn = isbn
        self.paginas = paginas
        self.edicion = edicion
        self.editorial = editorial
        self.ciudad = ciudad
        self.pais = pais
        self.fecha_edicion = fecha_edicio

    def get_titulo(self): return self.titulo
    def get_autor(self): return self.autor
    def get_isbn(self): return self.isbn
    def get_paginas(self): return self.paginas
    def get_edicion(self): return self.edicion
    def get_editorial(self): return self.editorial
    def get_ciudad(self): return self.ciudad
    def get_pais(self): return self.pais
    def get_fecha_edicion(self): return self.fecha_edicion

    def set_titulo(self, t): self.titulo = t
    def set_autor(self, a): self.autor = a
    def set_isbn(self, i): self.isbn = i
    def set_paginas(self, p): self.paginas = p
    def set_edicion(self, e): self.edicion = e
    def set_editorial(self, ed): self.editorial = ed
    def set_ciudad(self, c): self.ciudad = c
    def set_pais(self, pa): self.pais = pa
    def set_fecha_edicion(self, f): self.fecha_edicion = f

    def leer_informacion(self):
        return {
            "Título": self.titulo,
            "Autor": self.autor,
            "ISBN": self.isbn,
            "Páginas": self.paginas,
            "Edición": self.edicion,
            "Editorial": self.editorial,
            "Ciudad": self.ciudad,
            "País": self.pais,
            "Fecha de edición": self.fecha_edicion
        }

    def mostrar_informacion(self):
        return (f"Título: {self.titulo}\n"
                f"Autor: {self.autor}\n"
                f"ISBN: {self.isbn}\n"
                f"Páginas: {self.paginas}\n"
                f"Edición: {self.edicion}\n"
                f"Editorial: {self.editorial}\n"
                f"Lugar: {self.ciudad}, {self.pais}\n"
                f"Fecha de edición: {self.fecha_edicion}")
```
